### Trying to use nltk and manual feature engeneering

```python
def gender_features(word):
    return {'word': word[:]}
```

```python
#use of the articles name to classify the articles group
#use of manual feature engeneering

# Manual feature engineering
labeled_names = [(gender_features('Ladieswear'), 'female'), 
                 (gender_features('Lingeries/Tights'), 'neutral'),
                 (gender_features('Baby Sizes 50-98'), 'neutral'),
                 (gender_features('Menswear'), 'male'),
                 (gender_features('Ladies Accessories'), 'female'),
                 (gender_features('Sport'), 'neutral'),
                 (gender_features('Children Sizes 92-140'), 'neutral'),
                 (gender_features('Divided'), 'neutral'),
                 (gender_features('Children Sizes 134-170'), 'neutral'),
                 (gender_features('Children Accessories, Swimwear'), 'neutral')]

random.shuffle(labeled_names)

# The training set is used to train a new "naive Bayes" classifier.
classifier = nltk.NaiveBayesClassifier.train(labeled_names)

#classification of articles in the other array:
classified_data_one = []

for article in articles_group:
    classification = classifier.classify(gender_features(article))
    classified_data_one.append((article, classification))

# Print the classifications
for article, classification in classified_data_one:
    print(f"Article: {article}, Predicted Gender: {classification}")
```

```python
#definition of a feature extractor that works only with the last letter of the word
def gender_features(word):
    return {'last_letter': word[-3]}
```

```python
#trying to classify the data of the array above using the nltk library

#creation of a dictionary with words (plural and singular) that are associated with male, female, or neutral
labeled_names = [(gender_features('Lady'), 'female'), 
                 (gender_features('Ladies'), 'female'),
                 (gender_features('Girls'), 'female'),
                 (gender_features('Girl'), 'female'),
                 (gender_features('Men'), 'male'),
                 (gender_features('Mens'), 'male'),
                 (gender_features('Boy'), 'male'),
                 (gender_features('Boys'), 'male'),
                 (gender_features('Sport'), 'neutral'),
                 (gender_features('Divided'), 'neutral'),
                 (gender_features('Unisex'), 'neutral')]

random.shuffle(labeled_names)

#classification instance
classifier = nltk.NaiveBayesClassifier.train(labeled_names)

#classification of articles in the other array:
definitive_classification = []

for article in section_names:
    classification = classifier.classify(gender_features(article))
    definitive_classification.append((article, classification))

#print the classifications
for article, classification in definitive_classification:
    print(f"Article: {article}, Predicted Gender: {classification}")
```

```python
#trying to classify the data of the array above using the nltk library

#creation of a dictionary with words (only plural) that are associated with male, female, or neutral
labeled_names = [(gender_features('Ladies'), 'female'),
                 (gender_features('Girls'), 'female'),
                 (gender_features('Mens'), 'male'),
                 (gender_features('Boys'), 'male'),
                 (gender_features('Sport'), 'neutral'),
                 (gender_features('Divided'), 'neutral'),
                 (gender_features('Unisex'), 'neutral')]

random.shuffle(labeled_names)

#classification instance
classifier = nltk.NaiveBayesClassifier.train(labeled_names)

#classification of articles in the other array:
definitive_classification_plural = []

for article in section_names:
    classification = classifier.classify(gender_features(article))
    definitive_classification_plural.append((article, classification))

#print the classifications
for article, classification in definitive_classification_plural:
    print(f"Article: {article}, Predicted Gender: {classification}")
```

```python
#trying to classify the data of the array above using the nltk library

#creation of a dictionary with words (only singular) that are associated with male, female, or neutral
labeled_names = [(gender_features('Lady'), 'female'),
                 (gender_features('Girl'), 'female'),
                 (gender_features('Men'), 'male'),
                 (gender_features('Boy'), 'male'),
                 (gender_features('Sport'), 'neutral'),
                 (gender_features('Divided'), 'neutral'),
                 (gender_features('Unisex'), 'neutral')]

random.shuffle(labeled_names)

#classification instance
classifier = nltk.NaiveBayesClassifier.train(labeled_names)

#classification of articles in the other array:
definitive_classification_singular = []

for article in section_names:
    classification = classifier.classify(gender_features(article))
    definitive_classification_singular.append((article, classification))

#print the classifications
for article, classification in definitive_classification_singular:
    print(f"Article: {article}, Predicted Gender: {classification}")
```

```python
#phrases that have been classified in a different way in my three arrays
different_classified = []
equal_classified = []

for (phrase1, class1), (phrase2, class2), (phrase3, class3) in zip(definitive_classification, definitive_classification_plural, definitive_classification_singular):
    if class1 != class2 or class1 != class3 or class2 != class3:
        different_classified.append((phrase1, (class1, class2, class3)))
    else:
        equal_classified.append((phrase1, (class1, class2, class3)))
```

```python
#creation of the definitive classification using the class that is more present in the phrases that have been classified in different ways
most_common_classes = {}

for phrase, classes in different_classified:
    most_common_class = max(set(classes), key=classes.count)
    most_common_classes[phrase] = most_common_class

result_array = [(phrase, most_common_class) for phrase, most_common_class in most_common_classes.items()]

for phrase, most_common_class in result_array:
    print(f"Phrase: {phrase}, Most Common Class: {most_common_class}")
```

```python
#taking only one element for the class in the equal_classified array

class_to_use = {} 

for phrase, class_tuple in equal_classified:
    definitive_class = class_tuple[0]
    class_to_use[phrase] = definitive_class

result_array_equal = [(phrase, definitive_class) for phrase, definitive_class in class_to_use.items()]
```

```python
#array with the definitive classification for all the sentences

definitive_classification =[]
for element in result_array:
    definitive_classification.append(element)
for element_bis in result_array_equal:
    definitive_classification.append(element_bis)

definitive_classification
```
