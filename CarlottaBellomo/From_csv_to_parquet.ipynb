# Import libraries


```python
#articles
path_articles = 'gdrive/My Drive/Artificial_intelligence_project_UA/Data/articles.csv'
copyfile(path_articles, 'articles.csv')
articles = pd.read_csv('articles.csv')
```

```python
#customers
path_customers = 'gdrive/My Drive/Artificial_intelligence_project_UA/Data/customers.csv'
copyfile(path_customers, 'customers.csv')
customers = pd.read_csv('customers.csv')
```

```python
#transaction
path_transaction = 'gdrive/My Drive/Artificial_intelligence_project_UA/Data/transactions_train.csv'
copyfile(path_transaction, 'transactions_train.csv')
transaction = pd.read_csv('transactions_train.csv')
```

# Parquet file and save

```python
#new files
parquet_path = 'gdrive/My Drive/Artificial_intelligence_project_UA/Parquet_datasets/'
parquet_customers = 'customers.parquet'
parquet_articles = 'articles.parquet'
parquet_transactions_train = 'transactions_train.parquet'

#save of file
customers.to_parquet(parquet_path + parquet_customers)
articles.to_parquet(parquet_path + parquet_articles)
transaction.to_parquet(parquet_path + parquet_transactions_train)
```
