# Owned items over different timeframes

```python
import matplotlib.pyplot as plt

# Given data
weeks = [10, 5, 4, 3, 2, 1]
values = [0.01741, 0.01659, 0.01631, 0.01548, 0.01394, 0.01129]

# Plotting the data
plt.plot(weeks, values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Weeks')
plt.ylabel('Kaggle Score')
plt.title('Prediction Owned Items Over Different Timeframes')
plt.savefig(f"Repurchase_diff_times.png")

# Display the graph
plt.show()
```

# Popularity over different timeframes

```python
import matplotlib.pyplot as plt

# Timeframes and corresponding Kaggle scores
timeframes = ['Last Week', 'Last Month', 'Last Trimester','Last Half Year', 'Last Year',  'Last Year and Half', 'Last 2 Years']
kaggle_scores = [0.00738, 0.00564,0.00467, 0.0021, 0.00315, 0.00283, 0.00307]

# Plotting the data
plt.plot(timeframes, kaggle_scores, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Timeframe')
plt.ylabel('Kaggle Score')
plt.title('Kaggle Score On Popularity Over Different Timeframes')
plt.xticks(rotation=45, ha='right')

plt.savefig(f"pop_diff_times.png")
# Display the graph
plt.show()
```
