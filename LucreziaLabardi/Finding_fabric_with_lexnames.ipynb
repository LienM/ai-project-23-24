### Application to the sentences

```python
#apply to our sentences

#create a list with all the sentences
sentences = articoli["detail_desc"].unique().tolist()
#create an empty dictionary
fabrics= {}

for sentence in sentences:
  #for each sentence I tokenize and create an empty list
  words = nltk.word_tokenize(sentence)
  fab = []

  #for each word (token) in the sentence
  for word in words:
    #get the synset
    synsets = wn.synsets(word)
    for synset in synsets:
      #get the hypernyms
        hypernyms = synset.hypernyms()
        if hypernyms:
            for hypernym in hypernyms:
              #check if hypernyms are the type that we're looking for (related to fabric or material)
              if hypernym.name() in ["fabric.n.01", "knit.n.01", "create_from_raw_material.n.01", "solid.n.01", "material.n.01","chemical_element.n.01", "substance.n.01", "animal_skin.n.01", "swimsuit.n.01"]:
                fab.append(word)
                #create a list with all the words related to fabric and add to the dictionary for each sentence
                fabrics[sentence] = fab
```

```python
#function to apply material list to the sentences
def get_material_fab(frase):
  if frase in fabrics:
    return(fabrics[frase])
  else:
    return None

#create a new column with the words
articoli['fabric'] = articoli['detail_desc'].apply(get_material_fab)
```

```python
#we can try to add manually some words that are common
import re

def get_other_words(frase):
  new_words = []
  #words that starts with elastic
  start_elastic = re.findall(r'\belastic\w*\b', frase)
  if len(start_elastic)>0:
    new_words.append("elastic")

  #words that has knit inside
  has_knit = re.findall(r'\b\w*knit\w*\b', frase)
  if len(has_knit)>0:
    new_words.append("knit")

  words = nltk.word_tokenize(frase)
  # word 'microfibre' is not in the dictionary
  if "microfibre" in words:
    new_words.append("microfibre")

  if len(new_words) == 0:
    return False
  else:
    return new_words

#add the materials to existent arrays
def get_material_add(frase):
  other = get_other_words(frase)
  if other:
    if frase in fabrics:
      return fabrics[frase] + other
    else:
      return other
  else:
    if frase in fabrics:
      return fabrics[frase]
    else:
      return None

articoli['fabric_add'] = articoli['detail_desc'].apply(get_material_add)
```

```python
# remove stopwords and words with len<3 from fabrics
def get_stop(lista):
  nuova_lista = []
  for el in lista:
    if el not in stop_words:
      nuova_lista.append(el)
  for nuovo_el in nuova_lista:
    if len(nuovo_el)<3:
      nuova_lista.remove(nuovo_el)
  return nuova_lista

articoli['fab_non_stop'] = articoli['fabric_add'].apply(get_stop)
```

### Map fabrics to numbers

```python
# create a dictionary to map words to numbers
word_dict = {}
counter = 0

for lista in just_desc['fab_non_stop']:
  if type(lista) == list:
    for parola in lista:
        if parola not in word_dict:
            word_dict[parola] = counter
            counter += 1

# Apply to arrays of words, with -1 to nan and empty lists
def encode_lista(lista):
    if type(lista) != list:
        return [-1]
    return [word_dict.get(parola, -1) for parola in lista]

just_desc['fab_num'] = just_desc['fab_non_stop'].apply(encode_lista)
```

Checking if the words that we got are nouns or adjectives and if this selection increases too much the number of nan values.

Unsing both universal tagger and normal tagger.

```python
tessuti = {}
for index, row in articoli.iterrows():
    frase = row['detail_desc']
    tessuti[frase] = row['fab_non_stop']
```

```python
print(tessuti["jersey top with narrow shoulder straps."])
```

```python
#creating new dictionaries
fabrics_nouns = tessuti.copy()
fabrics_adj = tessuti.copy()
```

```python
print(fabrics_nouns["jersey top with narrow shoulder straps."])
print(fabrics_adj["jersey top with narrow shoulder straps."])
```

```python
# universal tagger

for frase in articoli["detail_desc"]:
    #tokenize every sentence
    tokens = word_tokenize(frase)
    #pos tagging with universal tagger
    pos_tags = pos_tag(tokens, tagset='universal')

  #creating list of nouns and adj in each sentence
    nouns = [word for word, pos in pos_tags if pos == 'NOUN']
    adjectives = [ad for ad, pos in pos_tags if pos == 'ADJ']

    #select the words
    for tessuto in fabrics_nouns[frase]:
      if tessuto not in nouns:
        fabrics_nouns[frase].remove(tessuto)

    for tessuto in fabrics_adj[frase]:
      if tessuto not in adjectives:
        fabrics_adj[frase].remove(tessuto)

```

```python
def get_nomi(frase):
  if frase in fabrics_nouns and fabrics_nouns[frase] != []:
    return(fabrics_nouns[frase])
  else:
    return None

def get_aggettivi(frase):
  if frase in fabrics_adj and fabrics_adj[frase] != []:
    return(fabrics_adj[frase])
  else:
    return None

#create a new column with the words
articoli['fabric_nouns'] = articoli['detail_desc'].apply(get_nomi)
articoli['fabric_adj'] = articoli['detail_desc'].apply(get_aggettivi)
```
