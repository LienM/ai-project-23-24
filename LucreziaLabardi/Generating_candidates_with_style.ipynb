### Creating customers style

```python
# getting just the customers who have done more than 5 transactions
num_purch = transactions['customer_id'].value_counts().reset_index()
num_purch.columns = ['customer_id', 'num_trans']

cust_more_5 = num_purch[num_purch['num_trans'] > 5]

trans_selected_cust = transactions[transactions['customer_id'].isin(cust_more_5['customer_id'])]

# Get for each customers the list of articles they bought
c2art = trans_selected_cust.groupby('customer_id')['article_id'].unique()
```

```python
c2art.head()
```

```python
%%time

import statistics

# initialize an empty dictionary
cust_style = {
    'customer_id': [],
    'fav_color': [],
    'fav_graph_app': [],
    'fav_type': [],
    'fav_fabric': []
}

for cus, arts in c2art.items():
    #getting articles ids bought by that customer
    art_selected_id = arts
    art_selected = art_style.loc[art_style['article_id'].isin(art_selected_id)]
    
    #getting the mode of each feature
    moda_color = statistics.mode(art_selected['perceived_colour_value_id'])
    moda_type = statistics.mode(art_selected['product_type_no'])
    moda_graph_app = statistics.mode(art_selected['graphical_appearance_no'])
        
    # getting favourite fabric (different because they're arrays not numbers)
    all_fab = np.array([])
    for fid in art_selected['fabric']:
        all_fab = np.concatenate((all_fab, fid))
    all_fab = all_fab.astype(int)
    
    moda_fabric = statistics.mode(all_fab)
    
    if moda_fabric > 0:
        # if the mode is an actual fabric use that
        cust_style["fav_fabric"].append(moda_fabric)
    else:
        #if the favourite is -1 (nan) try to filter and get an actual fabric
        filtered_values = all_fab[all_fab != -1]
        if len(filtered_values) > 0:
            # if we find one, use that
            mode_without_minus_one = statistics.mode(filtered_values)
            cust_style["fav_fabric"].append(mode_without_minus_one)
        else:
            #otherwise use -1 
            cust_style["fav_fabric"].append(-1)
    
    #append other info to the dictionary
    cust_style["customer_id"].append(cus)
    cust_style["fav_color"].append(moda_color)
    cust_style["fav_type"].append(moda_type)
    cust_style["fav_graph_app"].append(moda_graph_app)
    
```

```python
#from dictionary to dataframe
df_preferenze = pd.DataFrame(cust_style)
```

```python
%%time

import statistics

# initialize an empty dictionary
cold_cust_style = {
    'customer_id': [],
    'fav_color': [],
    'fav_graph_app': [],
    'fav_type': [],
    'fav_fabric': []
}

#for cold start users getting the most popular values as style
total_moda_color = statistics.mode(articles['perceived_colour_value_id'])
total_moda_type = statistics.mode(articles['product_type_no'])
total_moda_graph_app = statistics.mode(articles['graphical_appearance_no'])
total_all_fab = np.array([])
for fid in articles['fabric']:
    total_all_fab = np.concatenate((total_all_fab, fid))
total_all_fab = total_all_fab.astype(int)
total_moda_fabric = statistics.mode(total_all_fab)

#add to dictionary
for cold_cust in cust_less_5["customer_id"]:
    cold_cust_style["customer_id"].append(cold_cust)
    cold_cust_style["fav_color"].append(total_moda_color)
    cold_cust_style["fav_type"].append(total_moda_type)
    cold_cust_style["fav_graph_app"].append(total_moda_graph_app)
    cold_cust_style["fav_fabric"].append(total_moda_fabric)
    
```

```python
#transfrom into dataframe
cold_preferenze = pd.DataFrame(cold_cust_style)
cold_preferenze
```

Regarding following code, I though of evaluating by counting the percentage of items in the first 100 recommended, that the customers actually bought. 
Since it takes too long, I will need to find a way to reduce data and probably find a different way to evaluate.

```python
%%time

def find_matching_articles(customer_name, cc):
    #getting customer preferences
    customer_features = cc.loc[cc['customer_id'] == customer_name]  
    customer_info = customer_features.values.tolist()[0]
    customer_colour = customer_info[1]
    customer_graph = customer_info[2]
    customer_type = customer_info[3]
    customer_fabric = customer_info[4]

    #list of ordered matching articles
    matching_articles = []
    #scoring how much the article matches customers' style
    for index, row in a.iterrows():
        score = 0
        if row["product_type_no"] == customer_type:
            score += 1
        if row["graphical_appearance_no"] == customer_graph:
            score += 1
        if row["perceived_colour_value_id"] == customer_colour:
            score += 1
        if customer_colour in row["fabric"]:
            score += 1
        if score > 0:
            matching_articles.append((row["article_id"], score, row["price"]))

    #ordering by score and by price
    matching_articles.sort(key=lambda x: (-x[1], x[2]), reverse=False)
    first100 = matching_articles[:100]
    #get and return the first 100 
    return first100

def purchased_in_recommended(customer_name, first100):
    #how many articles of the 100 generated the customer actually bought
    num_purch = 0
    bought_items = grouped_trans[customer_name]
    for match in first100:
        art_id = match[0]
        if art_id in bought_items:
            num_purch += 1
    return num_purch

def get_perc_purch_in_recom(cc):
    #getting the percentage of articles bought by all customers among all the recommended
    total_cust = len(cc)
    total_purch = 0 
    
    for cust in cc["customer_id"]:
        matched_articles = find_matching_articles(cust, cc)
        purch_in_first100 = purchased_in_recommended(cust, matched_articles)
        total_purch += purch_in_first100
    
    print("Total percentage of purchased items among the first 100 recommended ones to customers are:")
    print(f"{total_purch/(total_cust*100)*100}")
    

get_perc_purch_in_recom(c200)        
```
